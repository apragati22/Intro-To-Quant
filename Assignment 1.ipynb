```python
def open_close(df,str):
  a=0
  b=0
  c=0
  d=0
  length=len(df)

  for i in range(0,length):
    t=df["Open"].iloc[i]
    s=df["Close"].iloc[i]
    ratio=t/s
    if(ratio>=0.35 and ratio<=0.5):
      a+=1
    elif(ratio>0.5 and ratio<=1):
      b+=1
    elif(ratio>1 and ratio<=3.5):
      c+=1
    elif (ratio>3.5):
      d+=1
  print("For the stocks of ",str," in the interval Jan 2022 to Jan 2023, the stocks Open/Close ratio was in between:")
  print("0.35 and 0.5: ",a,"times")
  print("0.5 and 1:  ",b,"times")
  print("1and 3.5:    ",c,"times")
  print("above 3.5:  ",d,"times")
  return
```

# Analysis of the stocks- "Google" and "Amazon" from Jan 2020 to Jan 2023:


The graph above shows the closing prices of Google and Amazon between Jan 2020 and Jan 2023.

We notice that both had a booming period from May 2020 to around September 2021.


























***Some reasons for the surge of Google stocks since 2020***:


*   YouTubes ad revenue was up about a third as people turned to online entertainment while they hunkered down at home to avoid the coronavirus.
*   Overall ad revenues for Google rose 10 percent despite the pandemics worsening in March.






*  Alphabet and Google CEO Sundar Pichai has revealed that videos in the new YouTube Shorts player and live video features are being liked by the public.

***Some reasons for the surge of Amazon stocks since 2020:***
*  More consumers relied on delivery of essential goods from Amazon during the Covid Pandemic and forced lockdown all over the world.

*   A huge boost in profit was largely driven from an investment in Rivian Automotive Inc and a price hike in its flagship Prime membership offering.
*   Amazons most profitable unit, the Amazon Web Services cloud-computing division, generated sales of 17.8 billion and operating profit of $5.29 billion, topping estimates.

*   While online store sales actually declined from last years pandemic-fueled gains, Amazons profitable cloud-computing and advertising businesses combined to more than makeup for it.


After September 2021, both began **declining**

***Some reasons for the plunge in Google's stocks after June 2022:***
*   Alphabet is struggling to compete with TikTok amid a broader economic downturn.


*   Alphabets other bets which include the Waymo self-driving car project, Wing drone delivery and Verily life sciences rose to great valuations but these moonshot projects produced a collective operating loss of $1.1 billion.




*   Nasdaq 100 Index jumped to its highest level in a month during regular trading and the Invesco QQQ exchange traded fund which caused major decline in Tech Stocks.

***Some reasons for the plunge of Amazon since 2022 are:***
*   It spent some of last year revenues reducing its warehouse workforce through attrition, and canceling or delaying plans to open warehouses in different parts of the country.


*   The Pandemic was receding and things were getting back to normal. So people would now step into offline stores to meet daily needs.
*   In November, it began what would be the largest set of job cuts among its corporate offices, expected to be over 18,000 jobs in total.



*   The company was seeing lower growth rates internationally as consumers abroad feel the effects of high inflation and the economic repercussions of war in Ukraine.
